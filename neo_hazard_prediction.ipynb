```python
import kagglehub
import os
path = kagglehub.dataset_download('ivansher/nasa-nearest-earth-objects-1910-2024')
print(path)
os.listdir(path)
```

```python

import pandas as pd
neo_path = (path+'/nearest-earth-objects(1910-2024).csv')
neo = pd.read_csv(neo_path)
neo.describe()
```

```python
neo.head()
```

```python
neo.columns
```

```python
neo = neo.drop('orbiting_body',axis=1)
```

```python
neo.columns
```

```python
neo = neo.dropna(axis=0)
```

```python
neo_en = neo.copy()
neo_en.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
object_col = ['is_hazardous']
neo_en[object_col] = oe.fit_transform(neo[object_col])
```

```python
neo_en.head()
```

```python
neo_en.isnull().values.sum()
```

```python
y=neo_en.is_hazardous
neo_features=['absolute_magnitude','estimated_diameter_min','estimated_diameter_max','relative_velocity','miss_distance']
```

```python
print('NOT hazard:  ' + str((y==0.0).sum()))
print('Is Hazard:  ' + str((y==1.0).sum()))
```

```python
x=neo_en[neo_features]
```

```python
x.describe()
```

```python
x.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=1)
model.fit(x,y)
```

```python
#making predictions for the first 5 neo's
print(model.predict(x.head()))
for i in range(6):
  print("Prediction: ","Hazardous" if model.predict(x.iloc[[i]])==1 else "Not Hazardous")
```

```python
#example of over_fitting
from sklearn.metrics import mean_absolute_error, classification_report
preds = model.predict(x)
print(mean_absolute_error(y,preds))
print(classification_report(y,preds))
```

```python
#importance of splitting data into train and test datasets
from sklearn.model_selection import  train_test_split
x_train,x_val,y_train,y_val=train_test_split(x,y,random_state=0)
```

```python
from sklearn.metrics import accuracy_score
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
val_preds = model.predict(x_val)
report1 = str(accuracy_score(y_val,val_preds)*100)
print("MAE: " + str(mean_absolute_error(y_val,val_preds)))
print(classification_report(y_val,val_preds))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfmodel=RandomForestClassifier(random_state=2)
rfmodel.fit(x_train,y_train)
val_preds = rfmodel.predict(x_val)
report2 = str(accuracy_score(y_val,val_preds)*100)
print("MAE: " + str(mean_absolute_error(y_val,val_preds)))
print(classification_report(y_val,val_preds))
```

```python
from xgboost import XGBClassifier
xgmodel=XGBClassifier(n_estimators=1500)
xgmodel.fit(x_train,y_train)
val_preds = xgmodel.predict(x_val)
report3 = str(accuracy_score(y_val,val_preds)*100)
print("MAE: " + str(mean_absolute_error(y_val,val_preds)))
print(classification_report(y_val,val_preds))
```

```python
print('DECISION TREE MODEL: '+ report1 + ' %')
print('RANDOM FOREST MODEL: '+ report2 + ' %')
print('XGBOOST MODEL: '+ report3 + ' %')
```
